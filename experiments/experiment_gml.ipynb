```python
# Function to calculate the area of a polygon given its vertices
def polygon_area(coords):
    """
    
    """
    n = len(coords)
    area = 0.0
    for i in range(n):
        j = (i + 1) % n
        area += coords[i][0] * coords[j][1]
        area -= coords[j][0] * coords[i][1]
    area = abs(area) / 2.0
    return area 
```
