```python
# Merge on builiding id is not possible, merge on groundsurfe ID # Needs to be extracted <bldg:GroundSurface gml:id="UUID_af71ad86-c132-46eb-a2a7-01f78ec1f25f">
import lxml.etree as ET
# from shapely.geometry import Polygon


# Define the namespaces used in the CityGML file
ns = {
        'core': "http://www.opengis.net/citygml/2.0",
        'gml': "http://www.opengis.net/gml",
        'bldg' : 'http://www.opengis.net/citygml/building/2.0',
        'energy' : 'http://www.sig3d.org/citygml/2.0/energy/1.0',
        # Add other necessary namespaces based on the file content
    }


file_path = r"C:\Users\felix\Programmieren\tecdm\data\examples\gml_data\partialMierendorffInselLoD2.gml"
groundsurface_list = []
# Parse the XML file
tree = ET.parse(file_path)
root = tree.getroot()
# Find all buildings and get the data 
for building in root.findall('.//{*}Building'):
    # Get all IDs and yoc 
    building_id = building.get('{http://www.opengis.net/gml}id')
    # 
    gs_node = building.find('.//{*}GroundSurface', ns)
    if gs_node is None:
            surface_id = ""
    else:
        surface_id =gs_node.get('{http://www.opengis.net/gml}id')
    groundsurface_list.append((building_id, surface_id))
```

```python
sheet_["groundsurfcace_id"] = sheet_["gml_id"].map(dict(groundsurface_list))
```
