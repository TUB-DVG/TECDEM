# This file contains the code, to generate the Simulation Sheet for the Mierendorffinsel 
It extract the  Building Level Data from the CityGML files and creates a subset, which is within the boundaries of the Mierendoff Insel 


```python
import lxml.etree as ET
# from shapely.geometry import Polygon


# Define the namespaces used in the CityGML file
ns = {
        'core': "http://www.opengis.net/citygml/2.0",
        'gml': "http://www.opengis.net/gml",
        'bldg' : 'http://www.opengis.net/citygml/building/2.0',
        'energy' : 'http://www.sig3d.org/citygml/2.0/energy/1.0',
        # Add other necessary namespaces based on the file content
    }
```

```python
# Needs to be extracted <bldg:GroundSurface gml:id="UUID_af71ad86-c132-46eb-a2a7-01f78ec1f25f">
file_path = r"C:\Users\felix\Programmieren\tecdm\data\examples\gml_data\partialMierendorffInselLoD2.gml"
groundsurface_list = []
# Parse the XML file
tree = ET.parse(file_path)
root = tree.getroot()
# Find all buildings and get the data 
for building in root.findall('.//{*}Building'):
    # Get all IDs and yoc 
    building_id = building.get('{http://www.opengis.net/gml}id')
    # 
    gs_node = building.find('.//{*}GroundSurface', ns)
    if gs_node is None:
            surface_id = ""
    else:
        surface_id =gs_node.get('{http://www.opengis.net/gml}id')
    groundsurface_list.append((building_id, surface_id))
```

```python
# Map the GML IDs to the blknr 
import pandas as pd
gml_bldg = pd.read_csv(r"C:\Users\felix\Programmieren\tecdm\data\berlin\09_GML_blocka.csv")
```

```python
gml_district = gml_bldg.loc[gml_bldg["gmlid"].isin([uid for bldg, uid in groundsurface_list])]
```

```python
tumples_as_df = pd.DataFrame(groundsurface_list, columns=["gml_bldg_id", "gmlid"])
# gml_district["gml_bldg_id"] = gml_district["gmlid"].map(dict(groundsurface_list))
```

```python
gml_district = gml_district.merge(tumples_as_df, on="gmlid", how="left")
```

```python
gml_district.drop(columns=['id', 'gmlid_code', 'name', 'name_codes', 'descriptio',
       'creation_d', 'terminatio', 'relative_t', 'relative_1', 'last_modif',
       'updating_p', 'reason_for', 'lineage', 'building_i', 'bez',
       'bezname', 'plr', 'blktypname', 'ewk', 'area', 'datum', ], inplace=True)
```
